```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("C:/Users/Jyoti/Downloads/StudentsPerformance.csv")



df.head()

```

```python
plt.figure()
plt.hist(df['math score'], bins=10)
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
plt.title("Distribution of Math Scores")
plt.show()
```

```python
avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

plt.figure()
avg_scores.plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.title("Average Scores by Gender")
plt.show()
```

```python
avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

plt.figure()
avg_scores.plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.title("Average Scores by Gender")
plt.show()
```

```python
plt.figure()
plt.scatter(df['reading score'], df['writing score'])
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.title("Reading vs Writing Scores")
plt.show()
```

```python
prep_avg = df.groupby('test preparation course')['math score'].mean()

plt.figure()
prep_avg.plot(kind='bar')
plt.xlabel("Test Preparation Course")
plt.ylabel("Average Math Score")
plt.title("Math Score vs Test Preparation")
plt.show()
```

```python
plt.figure()
plt.scatter(df['math score'], df['reading score'])
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.title("Math vs Reading Scores")
plt.show()
```

```python
plt.figure()
df['gender'].value_counts().plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.title("Gender Distribution")
plt.show()
```

```python
avg = df[['math score','reading score','writing score']].mean()

plt.figure()
plt.plot(avg)
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.title("Overall Average Student Performance")
plt.show()
```

```python
prep_scores = df.groupby('test preparation course')[['math score','reading score','writing score']].mean()

plt.figure()
prep_scores.plot(kind='bar')
plt.xlabel("Test Preparation Course")
plt.ylabel("Average Score")
plt.title("Scores vs Test Preparation Course")
plt.show()
```

```python
plt.figure()
df['lunch'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Lunch Type Distribution")
plt.ylabel("")
plt.show()
```

```python
edu_avg = df.groupby('parental level of education')[['math score','reading score','writing score']].mean()

plt.figure()
edu_avg.plot(kind='bar')
plt.xlabel("Parental Level of Education")
plt.ylabel("Average Score")
plt.title("Average Scores vs Parental Education")
plt.show()
```

```python
plt.figure()
df.boxplot(column='math score', by='gender')
plt.title("Math Score by Gender")
plt.suptitle("")
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()
```
